## CircleMarker

```python
# Instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# Loop through the 200 crimes and add each to the incidents feature group
with data_engine.connect() as conn:
    resp = conn.execute(sql_str)
    for lat, lng in resp.yield_per(1):
        incidents.add_child(
            folium.CircleMarker(
                [lat, lng],
                radius=7, # define how big you want the circle markers to be
                color='yellow',
                fill=True,
                fill_color='red',
                fill_opacity=0.4
            )
        )

# Add incidents to map
huzhou_map = folium.Map(location=[latitude, longitude], zoom_start=12, tiles='Stamen Toner')
huzhou_map.add_child(incidents)
```
